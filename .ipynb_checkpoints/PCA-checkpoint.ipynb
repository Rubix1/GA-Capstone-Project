```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shap as sp

from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('./online_shoppers_intention.csv')
```

```python
df.head()
```

```python
df.Weekend = df.Weekend.apply(lambda x: 1 if x else 0,)
```

```python
df.head()
```

```python
df.VisitorType.unique()
```

```python
df['VisitorType'].groupby(df['Revenue']).value_counts(normalize=True)
```

```python
y = df.Revenue
df.drop('Revenue',inplace=True,axis=1)
```

```python
df_dum = pd.get_dummies(df,drop_first=True)
```

```python
df_dum.head()
```

```python
scaler = MinMaxScaler()
```

```python
df_scaled = scaler.fit_transform(df_dum)
```

```python
pca = PCA()
pca.fit(df_scaled)
```

```python
pca_ratio = pca.explained_variance_ratio_
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(np.cumsum(pca_ratio))
ax.set_title('Customer Intention Dataset Explained Variance')
ax.set_xlabel('Number of Components')
ax.set_ylabel('Variance (%)')
plt.show()
```

```python
def horn_parallel_analysis(shape, iters=1000, percentile=95):
    pca = PCA(n_components=shape[1])
    eigenvals = []
    
    for i in range(iters):
        rdata = np.random.normal(0,1,size=shape)
        pca.fit(rdata)
        eigenvals.append(pca.explained_variance_)
    
    eigenvals = np.array(eigenvals)
    return np.percentile(eigenvals, percentile, axis=0)
```

```python
df_pa = horn_parallel_analysis(df_scaled.shape)
df_pa
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(list(range(1, df_dum.shape[1]+1)), pca.explained_variance_, lw=2, marker='o')
ax.plot(list(range(1, df_dum.shape[1]+1)), df_pa, lw=2, color='darkred', marker='o')

ax.set_title("Horns parallel analysis on Customer Intent principal components")
ax.set_xlabel("Principal Component")
ax.set_ylabel("Eigenvalue")

plt.legend(['Actual PCA Variance Explained', "Random Gaussian PCA Variance Explained"])
plt.show()
```

```python
pca = PCA(n_components=16)
pca.fit(df_scaled)
```

```python
df_pcs = pca.transform(df_scaled)
df_pcs = pd.DataFrame(df_pcs, 
                        columns=['PC'+str(i+1) for i in range(16)])
```

```python
df_pcs.head()
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import cross_val_score
```

```python
lgbm = LGBMClassifier()

lgbm.fit(df_pcs, y)
```

```python
score = cross_val_score(lgbm, df_pcs, y, cv=10)
```
