```python
from sklearn.decomposition import PCA
```

```python
df_cval = df.select_dtypes(float)
```

```python
df_cval.describe()
```

```python
df_cval.SpecialDay.unique()
```

```python
duration_features = df_cval.iloc[:,:3]
```

```python
pca = PCA()
dur_feat_pca = pca.fit(duration_features)
```

```python
dur_feat_pca.explained_variance_ratio_
```

```python
pca_feature = dur_feat_pca.transform(duration_features)
```

```python
pca_feature
```

```python
pca_feature = pd.DataFrame(pca_feature)
```

```python
pca_feature
```

```python
pca_feature = pca_feature.iloc[:,:1]
```

```python
pca_feature.rename(columns={'0':'pca_feature'}, inplace=True)
```

```python
df_mod_pca = df.join(pca_feature)
```

```python
df_mod_pca.drop(['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration'],inplace=True,axis=1)
```

```python
df_mod_pca
```

```python
df_mod_pca.describe()
```

```python
y_pca = df_mod_pca.Revenue
df_mod_pca.drop('Revenue',axis=1,inplace=True)
```

```python
from lightgbm import LGBMClassifier
```

```python
df_mod_pca = pd.get_dummies(df_mod_pca,drop_first=True)
```

```python
lbm = LGBMClassifier()
lbm.fit(df_mod_pca,y_pca)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score =  cross_val_score(lbm,df_mod_pca,y_pca,cv=10)
score.mean()
```
