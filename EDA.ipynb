```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv('./online_shoppers_intention.csv')
```

```python
dfprofile = ProfileReport(df, title='Pandas Profiling Report', html={'style':{'full_width':True}})
```

```python
#Histogram Analysis between Weekend and Revenue

df1 = pd.crosstab(df.Weekend,df.Revenue)
df1.div(df1.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True, figsize = (5, 4))
plt.title('Weekend vs Revenue', fontsize = 20)
plt.show()
```
